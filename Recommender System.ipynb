**Importing of the necessary libraries/modules**

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

The TfidVectorizer converts text data into umerical values. 

The cosine_similarity is used to find the similarity score of all the movies or the similarity score for all other data similar to the user input.

**Dowloading of the movies dataset into a pandas dataframe using the pandas module**

```python
# loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv("C:\\Users\\hp\\Downloads\\movies.csv")
```

**Select Relevant Columns**

Select the columns relevant to the recommender system. The genres, keywords, overview, tagline, title, casts, and director column names was selected and saved in a list with variable `selected_features`

```python
# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

The null values present in the selected columns were replaced with null strings

```python
# replacing the null valuess with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

**Combine selected columns**

Combine the selected columns in a string format, separating each row by its index

```python
# combining all the 5 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

**Convert text data to numerical data**

Using the Vectorizer module, convert the selected combined columns from a string data type to numerical values

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

**Similarity Score**

Used the cosine_similarity module to find the similarity score of the movies. 

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting the movie name from the user

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

The title of the movies will be the user iput, so create a list of all the movies present i the dataset using their title.

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

**Get close matches**

Used the difflib module to get the close match for the user input.

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

Find the index of the titles

```python
# finding the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

**Similarity Score**

Check for the similarity score of the user input movie (My date with drew) when compared to other movies on the movies dataset.

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

**Sort Similarity Score**

Sort the similarity score for the user input from highest to lowest (in descendig order) using the sorted() function

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

Print the name of the first 20 movies most similar to the user input

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

**The Recommender System**

The recommender system can be ru i oe go by running the various lines of code in one cell.

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
